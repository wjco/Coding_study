**[문제]**

루트 없는 트리가 주어진다. 이때, 트리의 루트를 1이라고 정했을 때, 각 노드의 부모를 구하는 프로그램을 작성하시오.

**[입력]**

첫째 줄에 노드의 개수 N (2 ≤ N ≤ 100,000)이 주어진다. 둘째 줄부터 N-1개의 줄에 트리 상에서 연결된 두 정점이 주어진다.

**[출력]**

첫째 줄부터 N-1개의 줄에 각 노드의 부모 노드 번호를 2번 노드부터 순서대로 출력한다.

```python
# My Solution (1) - DFS / 인접 행렬을 이용한 풀이 but "메모리 초과"

def dfs(now, links, visited, parents):
    visited[now] = 1
    for i in range(n):
        check = (links[now][i] == 1) and (visited[i] == 0)
        if check:
            parents[i] = now + 1
            dfs(i, links, visited, parents)
    return parents

n = int(input())
links = [[0 for _ in range(n)] for _ in range(n)]
parents = [0 for _ in range(n)]

for _ in range(n - 1):
    a, b = map(int, input().split())
    links[a - 1][b - 1] = 1
    links[b - 1][a - 1] = 1

parents = dfs(0, links, [0 for _ in range(n)], parents)

for p in parents[1:]:
    print(p)
```

```python
# My Solution (2) - DFS / 인접 리스트를 이용한 풀이

######## 백준에서 재귀 깊이 늘려주는 부분 ########

# import sys
# sys.setrecursionlimit(10**6)

##################################################

def dfs(now, links, visited, parents):
    visited[now] = 1
    for i in links[now + 1]:
        if visited[i - 1] == 0:
            parents[i - 1] = now + 1
            dfs(i - 1, links, visited, parents)
    return parents

n = int(input())
links = {i + 1: [] for i in range(n)}
parents = [0 for _ in range(n)]

for _ in range(n - 1):
    a, b = map(int, input().split())
    links[a].append(b)
    links[b].append(a)

parents = dfs(0, links, [0 for _ in range(n)], parents)

for p in parents[1:]:
    print(p)
```
